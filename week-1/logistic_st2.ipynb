```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
# loading data 
train_df = pd.read_csv("trainst2.csv")
test_df = pd.read_csv("testst2.csv")
```

```python
# adding non linear terms and training
X_train = train_df[['x', 'y']]
y_train = train_df['color']  

X_test = test_df[['x', 'y']]
y_test = test_df['color']

model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# final results and display 
print("Accuracy:", accuracy_score(y_test, y_pred)*100, "%")
plt.figure(figsize=(8, 6))
plt.scatter(X_test['x'], X_test['y'], c=y_pred, cmap='tab10', s=20, edgecolors='k', alpha=0.7)
plt.title("Predicted Classes - Logistic Regression")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
