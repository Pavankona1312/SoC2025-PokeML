```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
# loading data and splitting the labels
train_df = pd.read_csv("trainst1.csv")
test_df = pd.read_csv("testst1.csv")

X_train = train_df[['x', 'y']].copy()
y_train = train_df['color']  

X_test = test_df[['x', 'y']].copy()
y_test = test_df['color']
```

```python
# adding non linear terms and training
for df in [X_train, X_test]:
    df['x2'] = df['x'] ** 2
    df['xy'] = df['x'] * df['y']

model = LogisticRegression(max_iter=1000) 
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
# final results
print("Accuracy:", accuracy_score(y_test, y_pred)*100, "%")
plt.figure(figsize=(8,6))
plt.scatter(X_test['x'], X_test['y'], c=y_pred, cmap='bwr', edgecolors='k', s=20, alpha=0.7)
plt.title("Logistic Regression - Predicted Classes")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
